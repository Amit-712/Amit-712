```python
import pandas as pd
import seaborn as sns
df=pd.read_csv("https://github.com/AshishJangra27/Data-Analysis-with-Python-GFG/raw/main/20.%20Black%20Friday%20-%20Walkthrough/BlackFriday.csv")
```

```python
df.head()
```

```python
df['User_ID'].nunique()
del df['Product_Category_2']
del df['Product_Category_3']
```

```python
data = pd.DataFrame({'Ratio' : [len(df[df['Gender']=='M']) , len(df[df['Gender']=='F'])]} , index=["Male","Female"])
data.plot.pie(y = 'Ratio' , autopct='%.2f')
```

```python
df.groupby('Gender').size().plot(kind='pie' , autopct='%.2f' , title='Gender ratio')
```

```python
df.groupby('Gender').size().plot(kind='bar',title='Gender ratio')
```

```python
df.groupby('Gender').sum()['Purchase']
```
